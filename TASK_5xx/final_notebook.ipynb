```python
import pandas as pd

df = pd.DataFrame({
    "subject": ["Math","Science","Math","English","Science","English"],
    "score": [85,90,95,80,85,88]
})

# Correct: compute average per subject
avg_score_per_subject = df.groupby("subject")["score"].mean()
best_subject = avg_score_per_subject.idxmax()

print("Average score per subject:")
print(avg_score_per_subject)
print("Best subject:", best_subject)
```

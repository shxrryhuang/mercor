```python
#!/usr/bin/env python
# coding: utf-8

import numpy as np

data = [10, 12, 13, 15, 18, 21, 23, 25, 27, 30]
print("Original data:", data)

# Correct median
median_value = np.median(data)

# Correct MAD = median(|x - median|)
mad = np.median(np.abs(np.array(data) - median_value))

# Normalize using MAD
normalized = [(x - median_value) / mad for x in data]

print("Median:", median_value)
print("MAD:", mad)
print("Normalized data:", normalized)
```

```python
import pandas as pd
import numpy as np

# Product quality by factory
df = pd.DataFrame({
    "factory": ["A","A","A","B","B","B","C","C","C"],
    "quality_score": [80,82,85,78,77,79,90,92,91]
})

# Correct: compute std per factory
factory_stats = df.groupby("factory")["quality_score"].std()
highest_std_factory = factory_stats.idxmax()
highest_std_value = factory_stats.max()

print("Factory std devs:")
print(factory_stats)
print("Highest variability factory:", highest_std_factory)
```

```python
import pandas as pd
import numpy as np

# Create internal dataset
df = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'revenue': [1200, 1350, 1600, 1550, 1700, 1800, 2000, 1950, 2100, 2200, 2300, 2500]
})

def customer_trends():
    # Compute 3-month moving average
    moving_average = df['revenue'].rolling(window=3).mean().iloc[-1]

    # Predict next month's revenue using linear regression trend
    x = np.arange(len(df))
    y = df['revenue'].values
    coeffs = np.polyfit(x, y, 1)
    next_month = len(df)
    prediction = np.polyval(coeffs, next_month)

    result = {
        'moving_average': float(round(moving_average, 2)),
        'prediction': float(round(prediction, 2))
    }
    return result

# Example output
print(customer_trends())
```

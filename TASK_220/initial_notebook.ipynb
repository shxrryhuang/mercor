```python
import pandas as pd
import numpy as np

# Create internal dataset
df = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'revenue': [1200, 1350, 1600, 1550, 1700, 1800, 2000, 1950, 2100, 2200, 2300, 2500]
})

def customer_trends():
    # INCORRECT: using mean of all revenue instead of rolling 3-month mean
    moving_average = df['revenue'].mean()

    # Predict next month's revenue using linear regression
    x = np.arange(len(df))
    y = df['revenue'].values
    coeffs = np.polyfit(x, y, 1)
    next_month = len(df)
    prediction = np.polyval(coeffs, next_month)

    # Wrong keys â€” intentional
    result = {
        'avg': float(round(moving_average, 2)),
        'pred': float(round(prediction, 2))
    }
    return result

print(customer_trends())
```

```python
#!/usr/bin/env python
# coding: utf-8

data = [
    ("A", 80, 2),
    ("B", 70, 3),
    ("A", 90, 1),
    ("C", 85, 2),
    ("A", 75, 4)
]

category = "A"

# Filter only selected category
filtered = [(score, weight) for cat, score, weight in data if cat == category]

# Weighted average
weighted_avg = sum(score * weight for score, weight in filtered) / sum(weight for _, weight in filtered)

print("Weighted average for category A:", weighted_avg)
```

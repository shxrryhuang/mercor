```python
import pandas as pd
import numpy as np

data = {
    'city': ['NYC','LA','Chicago','Miami','Dallas'],
    'oct_temp':[60,75,55,82,70],
    'nov_temp':[52,72,48,78,64],
    'dec_temp':[40,68,35,75,58]
}

df = pd.DataFrame(data)
df['mean_temp'] = df[['oct_temp','nov_temp','dec_temp']].mean(axis=1)

print("Average City Temperatures:")
print(df)
```
